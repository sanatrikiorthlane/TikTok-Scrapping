## Extract localisation

```python
def extract_localisation(descriptions):
    
    Localisations = ['france', 'paris','Ãle-de-France','cannes','lyon','brussels','nantes', 'bruxelles', 'marseille', 'marseille', 'nice', 'toulouse', 'bordeaux', 'caen', 'lille','franÃ§aise']
    liste_de_chaine = []
    descrip = " "
    loc_cherchÃ©e = " "
    localisation = []
    for L in descriptions:
        descrip = L
        liste_de_chaine = descrip.split()
    
        for e in liste_de_chaine:
            for k in Localisations:
                if (e == k):
                    loc_cherchÃ©e = e
                    localisation.append(loc_cherchÃ©e)
        if (loc_cherchÃ©e == " "):
            localisation.append(" ")
        loc_cherchÃ©e = " "
        
    return localisation
```

## Convert subscribers to Float

```python
def convert_subscribers_float(abonnÃ©es):
    
    Liste = ['M', 'K']
    final_abonnÃ©e = []
    number = " "
    for n in abonnÃ©es:
        if ('K' in n or 'M' in n) :
            number = n
            for i in range(0, len(number)):
                if (number[i] == 'K'):
                    abonnÃ© = float(number[:i])
                    abonnÃ© = abonnÃ© * 1000
                    final_abonnÃ©e.append(abonnÃ©)
                elif (number[i] == 'M'):
                    abonnÃ© = float(number[:i])
                    abonnÃ© = abonnÃ© * 1000000
                    str_abonnÃ©e = str(abonnÃ©)
                    final_abonnÃ©e.append(abonnÃ©)
        else:
                    n_float = float(n)
                    final_abonnÃ©e.append(n_float)
                    
    return final_abonnÃ©e
```

## Convert likes to Float

```python
def convert_likes_float(likes):
    
    Liste = ['M', 'K']
    final_like = []
    number = " "
    for n in likes:
        if ('K' in n or 'M' in n) :
            number = n
            for i in range(0, len(number)):
                if (number[i] == 'K'):
                    like = float(number[:i])
                    like = like * 1000
                    final_like.append(like)
                elif (number[i] == 'M'):
                    like = float(number[:i])
                    like = like * 1000000
                    final_like.append(like)
        else:
                    n_float = float(n)
                    final_like.append(n_float)
    
    return final_like
```
